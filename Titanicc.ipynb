```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
```

```python
data = pd.read_csv(r"D:\SIM\Elevvo\titanic\train.csv") 
data.head()
```

```python
data.info()
```

```python
#check for missing values
data.isnull().sum()
```

```python
#check for duplicate values
len(data[data.duplicated()])
```

```python
data.describe()
```

```python
#filling or dropping null values
data['Age']=data['Age'].fillna(data['Age'].mean())
data.drop(columns=['Cabin'], inplace=True) 
data.isnull().sum()
```

```python
#visulaizing age count
plt.figure(figsize=(8,4))
sns.histplot(data=data, x='Age', hue='Survived', kde=True)
plt.title('Age Count by Survival')
plt.show()
```

```python
#visualize survival by gender
sns.barplot(x='Sex', y='Survived', data=data)
plt.title("Survival by Gender")
plt.show()
```

```python
#visualize survival by class
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title("Survival Rate by Class")
plt.show()
```

```python
updatedData=data.drop(['PassengerId','Parch','SibSp'], axis=1)
sns.heatmap(updatedData.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Heatmap for survival rates
survival_rates = data.pivot_table(index='Sex', columns='Pclass', values='Survived')
sns.heatmap(survival_rates, annot=True)
plt.title("Survival by Gender and Class")
plt.show()
```
